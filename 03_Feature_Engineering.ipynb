**Extract Brand from image name**

```python
data = pd.DataFrame(images, columns=['src']) # kolumnen von source-pfad ableiten
data['brand'] = data['src'].apply(lambda x : x.split('_')[0].split("/")[-1]) # _ als kolumnen split anerkennen etc. -> brand erkennen
data['brand'].value_counts().head(20) # Anzahl der car images pro brand ausgeben
```

```python
data = data[(data['brand']=='Volkswagen')|(data['brand']=='Toyota')|(data['brand']=='Ford')|(data['brand']=='BMW')|(data['brand']=='Nissan')|(data['brand']=='Audi')|(data['brand']=='Mercedes-Benz')|(data['brand']=='Honda')|(data['brand']=='Kia')|(data['brand']=='Volvo')].reset_index(drop=True)

data
```

# Feature engineering

**Change images to grey**

```python
X= []
y= []
for i in range(len(data)):
    src = data.loc[i,'src']
    src = cv2.imread(src, cv2.IMREAD_COLOR)
    dst = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    X.append(cv2.resize(dst, dsize=(200, 300), interpolation=cv2.INTER_AREA))
    y.append(data.loc[i,'brand'])
    
fig, axes = plt.subplots(1,10,figsize=(25,10))
for i in range(0,10):
    axes[i].imshow(X[i])
```

**Reshape array**

```python
X = np.array(X)
X = X.astype('float32')
X = X /255.0
X = X.reshape(-1,300,200,1)
y= np.array(pd.get_dummies(pd.DataFrame(y),columns=[0]))

X.shape, y.shape
```
