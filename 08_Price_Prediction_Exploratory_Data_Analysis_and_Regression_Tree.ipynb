### Modelling mit Decision Tree Regressor von Sklearn

```python
# Regression tree
from sklearn.tree import DecisionTreeRegressor # Decision Decision Tree Regressor
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics # Import scikit-learn metrics module for accuracy calculation


# Create Decision Tree classifer object
reg = DecisionTreeRegressor(max_depth=5)

# Train Decision Tree Classifer
reg = reg.fit(X_train, Y_train)

score = reg.score(X_test,Y_test)
print(score)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(reg, 
                out_file=dot_data,
                filled=True, 
                rounded=False,
                special_characters=True,
                feature_names = ['Make', 'Model', 'Year', 'Horsepower', 'Displacement', 'Engine Type', 'Drivetrain', 'Passenger Capacity'],
                class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('reg_tree.png')
Image(graph.create_png())
```
